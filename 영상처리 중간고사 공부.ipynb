## CH01

* 디지털 이미지 처리의 정의
디지털 이미지가 입/출력인 모든 형태의 디지털 처리 과정과 시스템 기술을 총칭
디지털 이미지가 입/출력인 모든 형태의 신호 처리 관련 기술,학문 등을 총칭
**협의**: 이미지에서 얻고자 하는 정보를 보다 유용하게 이용하고자 해당 이미지를 가공 및 변형시키는 제반 기술 및 이론
**광읜**: 협의를 포함하여, 이미지에 대한 "시각적 처리과정 및 해당 기능"의 이해 및 구현과 이미지로부터 필요한 정보 추출등에 관련된 제반 기술 및 이론

* computer vision과 Digital image processing의 차이
디지털 이미지 처리는 computer vision보다 이미지의 quality에 관심이 있다.
해당 이미지가 이용되는 작업의 목적에 맞춰 보다 이미지를 우수하게 만드는데 있다. 그래서 주로 이미지의 quality를 높이는 기법을 연구한다.
* CV는 주로 입력이 이미지이고 출력이 symbol(의미)이다. 이미지 처리는 입출력 모두 이미지

### Digital image
* 장점
1. 높은 재현성
2. 영상처리를 통한 다양한 정보 추출/추가 등이 쉬움
3. 취득, 처리,재현, 저장이 용이

* 단점
1. 데이터 손실

디지털 이미지 신호에는 아날로그(x-ray film)와 디지털 신호(Dicom,PACS)가 있다
** PACS (picture archiving and communication system)
의료용 영상 기기에서 얻어진 영상을 디지털 형태로 저장/관리
의료진 등의 사용자의 요구에 따라 의료영상을 조회/검색

### Digital image processing
* Digitization
1. Sampling(표본화): 원래의 영상정보(아날로그)에서 충분한 정보를 표현하면서도 제한된 용량으로 저장가능하도록 신호의 일부를 선별하는것. (비가역적)  

2. Quantization(양자화): 연속적인 아날로그값을 이산적인 값으로 바꾸는것
양자화 에러 존재
비가역적
계조도
3. Encoding(부호화): 양자화된 sample의 값을 이진수 등의 디지털 정보로 표현. 이 과정에서 보통 압축등이 이뤄지는 경우가 많다.

###  Digital image processing의 분류

* Image Enhancement
영상 개선
-영상으로부터 얻고자 하는 데이터를 보다 쉽게 획득하도록 영상을 개선하는것
(Contrast,spatial characteristics개선, 노이즈 제거, Sharpness(윤관을 뚜렷하게 처리)
Inverse: 눈은 밝은 배경보다 어두운 배경에서 밝기 변화에 민감함
Spatial Filtering
Sharpness(선예도): 이미지의 자세한 명확도를 나타냄 Acutance(첨예도): 얼마나 빠르게 이미지 정보rk edge로 바뀌는가. Resolution(해상도): 요소들의 디테일을 면밀하게 구별하는가
-주관적으로 화질 등을 향상시키는 기술을 말함
(포토샵 사용)

* Image Restoration
영상처리 과정에서 발생되는 노이즈나 왜곡을 제거하여 원래의 영상과 가까운 형태로 복원하는 기술. 수학적 모델에 기반한다는 특징이 주관적인 Image enhancement와 다름
1. 영상 화질을 객관적으로 향상 시키는 기술
2. 손상된 영상을 원본 영상으로 변환
3. 영상 훼손 원인을 모델링 후 역변환

* Image Compression
영상을 효율적으로 저장/전송하기 위해 불필요하거나 중복된 부분을 제거. 즉, 영상 정보를 효율적으로 표현하는 기술.
인코딩/디코딩으로 구성
무손실 압축은 가욕적 변환이나 손실압축은 비가역적 변환임

* Image Analysis
영상이 지닌 특징을 수치화하여 표현. 구조적 특징,통계적 특징 등을 추출. 추출된 특징만으로는 원 영상으로 복원 불가

* Image Synthesis
다른 영상이나, 영상이 아닌 데이터로부터 영상을 형성
